```python
#dictionary 
numbers = [1,2,3,4,5]
square_dict = dict()

for num in numbers:
  square_dict[num] = num**2
square_dict
```

```python
#dictionary comprehension
square_dict = {num:num**2  for num in numbers}
```

```python
square_dict
```
