```python
#calculator:
first_num = float(input("Enter your first num: "))
print("Press keys for operator: (+,-,*,/)")
operator = input("Enter your operator: ")
second_num = float(input("Enter your second num: "))

if operator == "+":
  print(first_num+second_num)
elif operator  == "-":
  print(first_num-second_num)
elif operator  == "*":
  print(first_num*second_num)
elif operator  == "/":
  print(first_num/second_num)  
else:
  print("Invalid operations. ")
```
