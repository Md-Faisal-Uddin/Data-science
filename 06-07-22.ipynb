```python
#nested loop
i = 1
while i<5: #outer loop
  j=1
  while j<5:
    print(j)
    j +=1
  i+=1 
```

```python
i = 1
while i<=5: #5rows
  j = 1
  while j<=i:
    print('*',end=' ')
    j = j+1
  print()
  i = i+1
```
