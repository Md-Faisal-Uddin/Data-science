**Class-11:Nested loop**

```python
rows = int(input("Enter rows: "))
symbol = input("Enter the symbol: ")

#outer loop
for x in range(1,rows+1):#initially x=0
  #inner loop
  for y in range(1,x+1):
    print(symbol, end=' ')
  print('')
```

```python
rows = int(input("Enter the row num: "))
columns = int (input("Enter the col num: "))
symbol = input("Enter a symbol: ")

#outer loop
for i in range(1,rows+1):
  #inner loop
  for j in range(1,columns+1):
    print(symbol, end = " ")
  print("")
```
